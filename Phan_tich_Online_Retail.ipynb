# 2. Xử lý dữ liệu khuyết thiếu

```python
na = df.isna()
na
```

```python
df[na.Description | na.CustomerID]
```

```python
# Phần trăm giá trị khuyết thuyết trong bộ dữ liệu gốc
df[na.Description | na.CustomerID].shape[0] / df.shape[0] * 100
```

- Nhận xét
    + Số dòng dữ liệu chứa giá trị khuyết thiếu chiếm tỷ trong lớn (~25%), nếu tiến hành xoá hết các dòng này sẽ dẫn đến phân tích sau này không còn chính xác nữa.
    + Ngoài ra dữ liệu khuyết thiếu chỉ tập trung vào hai trường giá trị là "Description" và "CustomerID", do đó nếu sử dụng phương thức df.fillna() để điền các giá trị thì tổng hợp và phân tích sau này cần xem lại cách đánh giá các số liệu liên quan đến hai trường kể trên.

```python
# Thay thế giá trị khuyết thiếu của thuộc tính Description bằng giá trị mặc định "Unknown"
values = {"Description": "Unknown", "CustomerID": 99999}
df.fillna(value = values, inplace = True)
```

```python
df.info()
```

# 3. Xử lý dữ liệu ngoại lai

```python
# Lọc bộ dữ liệu có thuộc tính Quantity >= 0 và UnitPrice > 0
condition_1 = df.Quantity >= 0
condition_2 = df.UnitPrice > 0
df_s = df[condition_1 & condition_2]
```

```python
df_s.head()
```

```python
df_s.info()
```

```python
df_s.shape[0] / df.shape[0] * 100
```
